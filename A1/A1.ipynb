```python
def func(factor, x):
    return np.exp(factor * x)


def derivative(func, factor, x, delta):
    return (1 / delta) * (
        (2 / 3) * (func(factor, x + delta) - func(factor, x - delta)) +
        (1 / 12) * (func(factor, x - 2 * delta) - func(factor, x + 2 * delta)))
```

```python
# let's evaluate the error at x=1
x = 1

delta = np.linspace(1.0e-9, 1.0, 100000)

err_e = np.abs(np.e - derivative(func, 1, x, delta))
err_e_001 = np.abs(np.exp(0.01) * 0.01 - derivative(func, 0.01, x, delta))

# finding optimal deltas
opt_a = delta[np.argmin(err_e)]
opt_b = delta[np.argmin(err_e_001)]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    delta,
    err_e,
    c="b",
    label="$f_a(x)=e^{x}$ | Optimal $\delta$ =" + " {:.2e}".format(opt_a),
)
plt.plot(
    delta,
    err_e_001,
    c="darkorange",
    label="$f_b(x)=e^{0.01x}$ | Optimal $\delta$ =" + " {:.2e}".format(opt_b),
)
plt.axvline(
    x=1.0e-3,
    c="b",
    ls="--",
    label="Rough Analytical Optimal $\delta$ for $f_a(x)=e^{x}$ ",
)
plt.axvline(
    x=1.0e-1,
    c="darkorange",
    ls="--",
    label="Rough Analytical Optimal $\delta$ for $f_b(x)=e^{0.01x}$ ",
)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Step size $\delta$")
plt.ylabel("Error $E(\delta)$")

plt.legend(loc="best")
```

We can see that our optimal deltas roughly appears around our predicted rough analytical values. Th error could be due to the approxamtion made for the $\epsilon_{i}$. # Q2: Lakeshore 670 Diodes

We want to write a routine that will take an arbitrary voltage and interpolate to return a temperature. 

We will interpolate T as a function of V using the Cubic Spline function from Scipy.interpolate. 

Note that in order to do so, we will need to reshape our data in such a way that V is a stricly increasing sequence and such that each voltage has its corresponding temperature at the same index position.

Also, we will assume that the voltage is given in Volt and the temperature in Kelvin.



```python
#This function will return the interpolated temperature associated with V_range as well as
# the interpolated temperature at desiered_V.


def get_T_inter(filename, desired_V):

    # loading the data
    data = np.loadtxt(filename)
    T = data[:, 0]
    V = data[:, 1]

    data_2d = np.vstack((V, T)).T
    # sorting by first collumn (V in this case)
    data_2d_shaped = data_2d[np.argsort(data_2d[:, 0])]

    V_shaped = data_2d_shaped[:, 0]
    T_shaped = data_2d_shaped[:, 1]

    # cubic spline interpolation
    cubic_spline = interpolate.CubicSpline(V_shaped, T_shaped)

    # evaluate the spline function
    V_range = np.linspace(V_shaped.min(), V_shaped.max(), 2000)
    T_interpolated = cubic_spline(V_range)
    print("The interporlated temperature at {} V is {:.3e} K.".format(
        desired_V, cubic_spline(desired_V)))
    return V_shaped, T_shaped, V_range, T_interpolated
```

```python
# T,V become their shaped version.
V, T, V_range, T_interpolated = get_T_inter("lakeshore.txt", desired_V=0.2)
```

```python
# plot the fit and data
plt.figure(figsize=(10, 6))
plt.plot(V, T, "ro", label="Data", markersize=5)
plt.plot(V_range, T_interpolated, label="Cubic Spline Interpolation", lw=3)
plt.xlabel("Voltage (V)")
plt.ylabel("Temperature (K)")
plt.legend()
plt.show()
```

Now we should make some sort of quantitative (but possibly rough) estimate of the error of our interpolation. 

To do so we will split our data set in two, by randomly choosing len(V)/2 points of our data sets. Let's call the set of randomly chosen points the training set and the remaining set of point the validation set. 
We will then refit the spline interpolation using the training set. The difference between the fit and the real T values of the validation set will give us an estimate of the error.

To have a "better" estimate of the error, it could maybe be a good idea to repeat the process more then once and take an average of the error estimation obtained at each iteration. This will give us an average error estimate of our interpolation.  

We will thus repeat this process 300 times, only showing the plots for the first and second interation.

```python
def err_approx(V, T, V_range, plot=True):
    # array of indices
    index = np.arange(len(V))

    # creating the trainign and validation set
    # replace = False : same item CANNOT be chosen more then once.
    train_index = np.sort(
        np.random.choice(index, int(len(index) / 2), replace=False))
    train_V, train_T = V[train_index], T[train_index]
    val_index = np.setdiff1d(index, train_index)
    val_V, val_T = V[val_index], T[val_index]

    # interpolation
    spline_train = interpolate.CubicSpline(train_V, train_T)
    T_train = spline_train(V_range)

    # error estimate
    err = val_T - spline_train(val_V)
    avg_err = np.mean(np.abs(err))

    if plot:
        plt.figure(figsize=(10, 6))
        plt.plot(train_V, train_T, "*", label="Training set data")
        plt.plot(val_V, val_T, "*", label="Validation set data")
        plt.plot(V_range, T_train, label="Cubic Spline Fit [Training set]")
        plt.xlabel("Voltage (V)")
        plt.ylabel("Temperature (K)")
        plt.legend(loc='best')
        plt.show()

        plt.figure(figsize=(10, 7))
        plt.plot(val_T, err, label="Residuals")
        plt.ylabel("Residuals (K)")
        plt.xlabel("Temperature (K)")
        plt.axhline(avg_err,
                    ls="--",
                    c="r",
                    label="Average Error = {:.2e}".format(avg_err))
        plt.legend(loc='best')
        plt.show()

    return avg_err
```

```python
err_arr = []
for i in range(300):
    if i == 0 or i == 1:
        err_arr.append(err_approx(V, T, V_range))
    else:
        err_arr.append(err_approx(V, T, V_range, plot=False))
print(
    'Thus, the "rough" average error on the interpolation is: {:.2e} K'.format(
        np.mean(err_arr)))
```

# Q3: Rational, Polynomial and Spline fit

```python
#Rational fit code from class
def rat_eval(p, q, x):
    top = 0
    for i in range(len(p)):
        top = top + p[i] * x**i
    bot = 1
    for i in range(len(q)):
        bot = bot + q[i] * x**(i + 1)
    return top / bot


def rat_fit(x, y, n, m, lin_func_inv):
    assert (len(x) == n + m - 1)
    assert (len(y) == len(x))
    mat = np.zeros([n + m - 1, n + m - 1])
    for i in range(n):
        mat[:, i] = x**i
    for i in range(1, m):
        mat[:, i - 1 + n] = -y * x**i
    pars = np.dot(lin_func_inv(mat), y)
    p = pars[:n]
    q = pars[n:]
    return p, q
```

```python
def multiple_inter_compa(n,
                         m,
                         start_x,
                         end_x,
                         func,
                         lin_func_inv,
                         title,
                         set_plot_lim=False,
                         print_pq=False):

    x = np.linspace(start_x, end_x, n + m - 1)
    y = func(x)
    xx = np.linspace(start_x, end_x, 1001)
    yy_true = func(xx)

    p, q = rat_fit(x, y, n, m, lin_func_inv)
    yy_rat = rat_eval(p, q, xx)
    if print_pq:
        print('p coef: {}'.format(p))
        print('q coef: {}'.format(q))

    pp_nm = np.polyfit(x, y, n + m)
    # Use same number of terms. Since using n+m results in a "Polyfit
    # may be poorly conditioned" warning we will also plot 2nd order.
    yy_poly_nm = np.polyval(pp_nm, xx)

    pp = np.polyfit(x, y, 2)
    yy_poly = np.polyval(pp, xx)

    cubic_spline = interpolate.CubicSpline(x, y)

    # evaluate the spline function

    yy_spline = cubic_spline(xx)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, "*", label="Data points")
    plt.plot(xx, yy_true, ls='-', c='darkorange', label="Truth")
    plt.plot(xx, yy_rat, ls='--', c='blue', label="Rational Fit")
    plt.plot(xx, yy_poly, ls='--', c='red', label="2nd order Polynomial Fit")
    plt.plot(xx,
             yy_poly_nm,
             ls='--',
             c='purple',
             label="(n+m) order Polynomial Fit")
    plt.plot(xx, yy_spline, ls='--', c='green', label="Spline Fit")
    plt.xlabel('x')
    plt.title(title)
    if set_plot_lim:
        plt.xlim(-1, 1)
        plt.ylim(-1, 10)

    plt.legend(loc='best')

    plt.figure(figsize=(10, 6))
    plt.xlabel('x')
    plt.ylabel('Residuals')
    plt.plot(xx,
             yy_rat - yy_true,
             ls='--',
             c='blue',
             label="Rational fit RMS error = {:.2e}".format(
                 np.std((yy_rat - yy_true))))
    plt.plot(xx,
             yy_poly - yy_true,
             ls='--',
             c='red',
             label="2nd order Polynomial Fit RMS error = {:.2e}".format(
                 np.std(yy_poly - yy_true)))
    plt.plot(xx,
             yy_poly_nm - yy_true,
             ls='--',
             c='purple',
             label="(n+m) order Polynomial Fit RMS error = {:.2e}".format(
                 np.std(yy_poly_nm - yy_true)))
    plt.plot(xx,
             yy_spline - yy_true,
             ls='--',
             c='green',
             label="Spline fit RMS error = {:.2e}".format(
                 np.std(yy_spline - yy_true)))
    plt.legend(loc='best')
    if set_plot_lim:
        plt.xlim(-1, 1)
        plt.ylim(-1, 10)
```

Note: We can notice that the (n+m) polynomial fit generates the warning "Polyfit may be poorly conditioned". It is because it is an illposed problem where a small change of our data set could result in a large change of the fitted polynomial.

Let's see the cosine case:

```python
n, m = 4, 7
multiple_inter_compa(n, m, -np.pi / 2, np.pi / 2, np.cos, np.linalg.inv,
                     'Fit for $y(x) = \cos(x)$, n = {}, m = {}'.format(n, m))
```

Now for the Lorentzian case, we expect the error from the rational fit to be zero as if p=[1] and q=[0, 1] then we get the exact same form as the Lorentizian. Which is exactly what we observe for n=1, m=3.

```python
def lorentz(x):
    return 1 / (1 + x**2)


n, m = 1, 3
multiple_inter_compa(n,
                     m,
                     -1,
                     1,
                     lorentz,
                     np.linalg.inv,
                     'Fit for $y(x)=1/(1+x^2)$ ' +
                     'n = {}, m = {}'.format(n, m),
                     set_plot_lim=False,
                     print_pq=True)
```

When we go to higher orders of n and m we are forcing the function to look less and less like the Lorentizian as we are introducing higher order terms. For example setting m=4, n=5 will have a tendency to make it act as 1/x close to zero as they differ by one degree. Looking at the plots we can see that this is happening as we have the asymptotic behaviour near x = 0. 

```python
n,m=4,5
multiple_inter_compa(n,
                     m,
                     -1,
                     1,
                     lorentz,
                     np.linalg.inv,
                     'Fit for $y(x)=1/(1+x^2)$ ' +
                     'n = {}, m = {}'.format(n, m),
                     set_plot_lim=True,
                     print_pq=True)
```

Going into higher order will tend to make the matrice singular (determinant very close to zero and thus no real inverse can be computed) which has the effect of introducing poles since we are dividing by 1/det(M). The p and q coefficient appears to be quite big compared to what they should be due to that singularity. Now, if we use np.linalg.pinv it will stop trying to compute the exact inverse of that matrix. It will approximate it (pseudo-inverse) and will allow p and q coefficent to be very close to zero. This behaviour is seen by looking at the coeficcient matrices p and q where the coeffcients that should be zero in order to have perfect fit are now of order e-15/e-16!

```python
multiple_inter_compa(n,
                     m,
                     -1,
                     1,
                     lorentz,
                     np.linalg.pinv,
                     'Fit for $y(x)=1/(1+x^2)$ ' +
                     'n = {}, m = {}'.format(n, m),
                     set_plot_lim=False,
                     print_pq=True)
```

We will try to evaluate this integral using on the one hand our recursive step integrator (simple_integrate) written in class and on the other hand scipy.intergrate.quad.

```python
def simple_integrate(fun, a, b, tol):
    x = np.linspace(a, b, 5)
    dx = (b - a) / 4.0
    #np.median(np.diff(x))
    y = fun(x)
    neval = len(x)  #let's keep track of function evaluations
    f1 = (y[0] + 4 * y[2] + y[4]) / 6.0 * (b - a)
    f2 = (y[0] + 4 * y[1] + 2 * y[2] + 4 * y[3] + y[4]) / 12.0 * (b - a)
    myerr = np.abs(f2 - f1)
    #print([a,b,f1,f2])
    if (myerr < tol):
        #return (f2)/1.0,myerr,neval
        return (16.0 * f2 - f1) / 15.0, myerr, neval
    else:
        mid = 0.5 * (b + a)
        f_left, err_left, neval_left = simple_integrate(fun, a, mid, tol / 2.0)
        f_right, err_right, neval_right = simple_integrate(
            fun, mid, b, tol / 2.0)
        neval = neval + neval_left + neval_right
        f = f_left + f_right
        err = err_left + err_right
        return f, err, neval
```

```python
# since we just want the function form so let sigma, epislon be 1.
epsilon = 1.  #constants.epsilon_0
sigma = 1.
R = 1


def E(z, R):
    return lambda theta: (sigma / (4 * np.pi * epsilon)) * (
        (z - R * (np.cos(theta))) * (2 * np.pi * R * np.sin(theta)) *
        (R)) / (z**2 + R**2 - 2 * z * R * (np.cos(theta)))**(3 / 2)


start = 0
end = 2
nb_points = 100


def integrate_E_field(start, end, nb_points, R, E_func):
    # let's create a z array with R inside
    z_left = np.linspace(start, R, nb_points, endpoint=False)
    z_right = np.linspace(R, end, nb_points, endpoint=False)
    z = np.concatenate((z_left, z_right))

    # array for integrated field
    E_quad, E_our, E_true = np.empty((3, len(z)))

    assert R in z
    assert (start < R and end > R)

    for i, zi in enumerate(z):
        if zi < R:
            E_true[i] = 0
        elif zi > R:
            E_true[i] = 1 / (4 * np.pi *
                             epsilon) * 4 * np.pi * R**2 * sigma / zi**2
        else:
            E_true[i] = np.nan

        E_true[i]
        try:
            E_quad[i] = integrate.quad(E_func(zi, R), 0, np.pi)[0]
        except:
            E_quad[i] = np.nan
            if zi == R:
                print(
                    "Error while handling the singularity at z = R (scipy.integrate.quad)"
                )
            else:
                print("Error while handling z = {} (scipy.integrate.quad)".
                      format(zi))
        try:
            E_our[i], _, _ = simple_integrate(E_func(zi, R), 0, np.pi, 1e-3)
        except:
            E_our[i] = np.nan
            if zi == R:
                print(
                    "Error while handling the singularity at z = R (Simple_Integrate)"
                )

            else:
                print("Error while handling z = {} (scipy.integrate.quad)".
                      format(zi))
    return E_quad, E_our, E_true, z
```

```python
E_quad, E_our, E_true, z = integrate_E_field(start, end, nb_points, R, E)
plt.figure(figsize=(10, 6))
plt.plot(z, E_quad, '--', label="Scipy.integrate.quad")
plt.plot(z, E_our, label="Simple_Integrate")
plt.plot(z, E_true, 'x', markersize=1.5, label="Analytic")
plt.xlabel("Distance from center of the sphere z [unit.]")
plt.ylabel("Electric Field E(z) [unit]")
plt.legend()

plt.figure(figsize=(10, 6))
plt.plot(z, E_quad, 'x', label="Scipy.integrate.quad")
plt.plot(z, E_our, label="Simple_Integrate")
plt.plot(z, E_true, 'x', markersize=1.5, label="Analytic")
plt.xlabel("Distance from center of the sphere z")
plt.ylabel("Electric Field E(z)")
plt.xlim([R - 0.1, R + 0.1])
plt.legend()
```

It is clear that there is a singularity in our integral at z = R (denominator will be equal to zero). 

Scipy.integrate.quad doesn't fail and handle the singularity. However our simple_integrate (recursive step integrator) function written in class fails at the singularity. Indeed, we get an "invalid value encountered in true_divide" as we are trying to divide by 0. We can also have a recursion error "maximum recursion depth exceeded while calling a Python object" as because of the singularity, we have an infinite loop and python realized we would never quit it.

We set the value of our simple_integrate function to be np.nan when z=R.
