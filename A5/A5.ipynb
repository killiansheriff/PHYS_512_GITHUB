```python
def get_strain_fft_ps(strain):
    strain_fft = np.fft.rfft(strain)
    ps = np.abs(strain_fft**2)
    return strain_fft, ps


def plot_ps_freq(data, labels, limits=[]):
    # data=[x1,y1,y2,y3,y4]
    fig, ax = plt.subplots(2, 1, sharex=True)
    ax[0].loglog(data[0],
                 data[1],
                 color='lightblue',
                 alpha=0.7,
                 label=labels[0])
    ax[0].loglog(data[0], data[2], color='orange', alpha=0.7, label=labels[1])

    ax[1].loglog(data[0],
                 data[3],
                 color='lightgreen',
                 alpha=0.7,
                 label=labels[2])
    ax[1].loglog(data[0], data[4], color='red', alpha=0.7, label=labels[3])

    if len(limits) > 1:
        ax[0].set_xlim(left=data[0][limits[0]], right=freqs[limits[1]])
    if len(data) > 5:  #adding smoothed data
        ax[0].loglog(data[0],
                     data[5],
                     alpha=1,
                     color='darkblue',
                     label=labels[4])
        ax[1].loglog(data[0],
                     data[6],
                     alpha=1,
                     color='darkgreen',
                     label=labels[5])

    ax[0].set_ylabel("Power")
    ax[1].set_xlabel("Frequency (Hz)")
    ax[1].set_ylabel("Power")
    ax[0].legend()
    ax[1].legend()


def get_windowed_data(strain, win_func):
    win = win_func(len(strain))
    strain_fft_win = np.fft.rfft(win * strain)
    ps_win = np.abs(strain_fft_win)**2
    return strain_fft_win, ps_win


def get_m_by_prewhite(N_inv_sqrt, strain_fft_win, template_fft_win, n):
    strain_fft_prewhite = N_inv_sqrt * strain_fft_win
    strain_prewhite = np.fft.irfft(strain_fft_prewhite, n)
    template_fft_prewhite = template_fft_win * N_inv_sqrt
    template_prewhite = np.fft.irfft(template_fft_prewhite, n)
    m_fft = strain_fft_prewhite * np.conj(template_fft_prewhite)
    m = np.fft.irfft(m_fft, n)
    return strain_prewhite, template_prewhite, template_fft_prewhite, m


def plot_m(data, labels):

    fig, ax = plt.subplots(2, 1, sharex=True)
    ax[0].plot(data[0], data[1], color='darkblue', label=labels[0])
    ax[1].plot(data[0], data[2], color='darkgreen', label=labels[1])

    ax[1].set_xlabel(labels[2])
    ax[0].set_ylabel(labels[-1])
    ax[1].set_ylabel(labels[-1])

    ax[0].legend()
    ax[1].legend()


def plot_SNR(data, labels):

    fig, ax = plt.subplots(3, 1, sharex=True)
    ax[0].plot(data[0], data[1], color='darkblue', label=labels[0])
    ax[1].plot(data[0], data[2], color='darkgreen', label=labels[1])
    ax[2].plot(data[0], data[3], color='darkorange', label=labels[2])

    ax[2].set_xlabel(labels[3])
    ax[0].set_ylabel(labels[-1])
    ax[1].set_ylabel(labels[-1])
    ax[2].set_ylabel(labels[-1])
    ax[0].legend()
    ax[1].legend()
    ax[2].legend()
```

# 1. a)

We want to come up with a noise model for Livingston and Hanford separately. To do so, we will calculate the power specturum of the strain data as it can be an estimate of our noise matrix N. It will be calcuted by taking the magnitude square of the FFT.

Since the data has been collected over a discrete period of time, the edges of our data set could be discountinous which may cause problem with the DFT. We will thus window our data by multiplying it by a window function that has an extended flat period near the center to avoid tapering the data/template where the signal is not small and of course one that goes to 0 at the edges. 
After looking at the different window function on wikipedia, we will choose the Tukey window. 

We then have to smooth the windowed power spectrum so that the features become less spiky. To do so, we will select a range of frequency in which the spectrum ressembles white noise, so a range where it is relatively of constant amplitude and randomly fluctuating. We then use a gaussian filter with std 50 to smooth the data and remove the spikes. 

After these steps, our Noise matrix N will be the 1D array of the power spectrum smoothed and windowed.
s we cut the data, for selecting a white noise behaving region, we set the values outside the range of frequencies to ∞ so that these regions do not contribute anymore (ie: will be 0 in the inverse). 

For this problem set, as example, we will use the event “GW150914”. Results are shown below and the ideal frequency (Hz) range for white noise behaviour was determined to be [20, 1650].

For conveniance, let define a naming convention: _h will be variable associated with hanford data, _l with livingston, _fft fourier transform, _ps power spectrum, _win windowed data.

```python
fold = "./LOSC_Event_tutorial/"
fnjson = fold + "BBH_events_v3.json"
eventname = 'GW150914'
#eventname = 'GW151226'
#eventname = 'LVT151012'
#eventname = 'GW170104'

event = json.load(open(fnjson, "r"))[eventname]
fn_h = fold + event['fn_H1']
fn_l = fold + event['fn_L1']
fn_template = fold + event['fn_template']

strain_h, dt_h, utc_h = read_file(fn_h)
strain_l, dt_l, utc_l = read_file(fn_l)
template_h, template_l = read_template(fn_template)
```

```python
if dt_h == dt_l and len(strain_h) == len(strain_l):
    dt, n = dt_h, len(strain_h)

sampling_rate = 1.0 / dt
freqs = np.fft.rfftfreq(n, d=dt)
```

```python
strain_h_fft, h_ps = get_strain_fft_ps(strain_h)
strain_l_fft, l_ps = get_strain_fft_ps(strain_l)
template_h_fft, template_h_ps = get_strain_fft_ps(template_h)
template_l_fft, template_l_ps = get_strain_fft_ps(template_l)
```

```python
data = [freqs, h_ps, template_h_ps, l_ps, template_l_ps]
labels = [
    "Hanford Data", "Hanford Template", "Livingston Data",
    "Livingston Template"
]

plot_ps_freq(data, labels)
```

```python
strain_h_fft_win, h_ps_win = get_windowed_data(strain_h,
                                               scipy.signal.windows.tukey)
strain_l_fft_win, l_ps_win = get_windowed_data(strain_l,
                                               scipy.signal.windows.tukey)
template_h_fft_win, template_h_ps_win = get_windowed_data(
    template_h, scipy.signal.windows.nuttall)
template_l_fft_win, template_l_ps_win = get_windowed_data(
    template_l, scipy.signal.windows.nuttall)
```

```python
data_win = [freqs, h_ps_win, template_h_ps_win, l_ps_win, template_l_ps_win]
labels_win = [
    "Hanford Data Windowed", "Hanford Template Windowed",
    "Livingston Data Windowed", "Livingston Template Windowed"
]
plot_ps_freq(data_win, labels_win)
```

```python
lim_down = np.argmin(np.abs(freqs - 20))
lim_up = np.argmin(np.abs(freqs - 1650))
lim = [lim_down, lim_up]
plot_ps_freq(data_win, labels_win, lim)
```

```python
h_ps_win_smooth = scipy.ndimage.gaussian_filter(h_ps_win, sigma=50)
l_ps_win_smooth = scipy.ndimage.gaussian_filter(l_ps_win, sigma=50)
```

```python
data_win_smooth = [
    freqs, h_ps_win, template_h_ps_win, l_ps_win, template_l_ps_win,
    h_ps_win_smooth, l_ps_win_smooth
]
labels_win_smooth = [
    "Hanford Data Windowed", "Hanford Template Windowed",
    "Livingston Data Windowed", "Livingston Template Windowed",
    "Hantford Data Smoothed", "Livingston Data Smoothed"
]

plot_ps_freq(data_win_smooth, labels_win_smooth, lim)
```

We can clearly see that the smoothed data has smaller fluctuations and the spikes have been reduced. The Tukey window function flatten this range that looks like white noise. The powerspectrum in that range for each detector and with infinity added outside the range becomes our noise model.

```python
N_h = h_ps_win_smooth
N_l = l_ps_win_smooth

N_h_inv_sqrt = 1 / np.sqrt(N_h)
N_l_inv_sqrt = 1 / np.sqrt(N_l)

N_h_inv_sqrt[:lim[0]], N_h_inv_sqrt[lim[1]:] = 0, 0
N_l_inv_sqrt[:lim[0]], N_l_inv_sqrt[lim[1]:] = 0, 0
```

# b)

We now want to use the above noise model to search for the sets of events using a matched filter.
As seen in class we have $m=(A^TN^{-1}d(A^TN^{-1}A)^{-1}$ with m the set of parameters, A the template data, d the strain data. If our noise does not behave like white noise, it won't be accurate so we need to pre-whitten the data! 




Splitting N in $N^{-1/2}$ factor, we have that our new noise matrix is the identity and thus behave like white noise: $m=((N^{-1/2}A)^T(N^{-1/2}d))((N^{-1/2}A)^T (N^{-1/2}A))^{-1}$. So we thus need to apply the transformation $A$ goes to $N^{-1/2}A$ and $d$ goes to $N^{-1/2}d$ so that $N$ goes to the identity matrix and thus behave like white noise.

We then proceed the the matched filter determination. We take the FFT of $N^{-1/2}d$ with the conjugate of $N^{-1/2}A$. $m$ will be the iFFT of this. 

```python
h_strain_prewhite, h_template_prewhite, h_template_fft_prewhite, m_h = get_m_by_prewhite(
    N_h_inv_sqrt, strain_h_fft_win, template_h_fft_win, n)
l_strain_prewhite, l_template_prewhite, l_template_fft_prewhite, m_l = get_m_by_prewhite(
    N_l_inv_sqrt, strain_l_fft_win, template_l_fft_win, n)
```

```python
time = np.arange(0, n) * dt
```

```python
data_filt = [time, np.abs(m_h), np.abs(m_l)]
labels_filt = ['Hanford', 'Time (s)', 'Matched Filter']

plot_m(data_filt, labels_filt)
```

# c) 

We now have to estimate a noise for each event, and from the output of the matched filter, to give a signal-to-noise ratio for each event, both from the individual detectors, and from the combined Livingston + Hanford events.

Similarly as what we seen with Least Square fitting, $Cov=A^TN^{-1/2}N^{-1/2}A=N^{-1}A^2$ since 1D arrays. The noise estimation will thus be the mean of our template pre-whitten square-rooted. The SNR is then obtain by diving the matched filter m by this estimated noise.

```python
sig_h = np.sqrt(np.mean(h_template_prewhite**2))
sig_l = np.sqrt(np.mean(l_template_prewhite**2))
#sig_comb = np.sqrt(np.mean((h_template_prewhite**2+l_template_prewhite**2)))
sig_comb = np.sqrt(sig_h**2 + sig_l**2)
print("Hanford noise estimate is {:.4e}".format(sig_h))
print("Livingston noise estimate is {:.4e}".format(sig_l))
print("Hanford + Livingston noise estimate is {:.4e}".format(sig_comb))
```
